```python
# Contoh dataset dalam bahasa Indonesia
data = {
    "intents": [
        {"tag": "salam", "patterns": ["Hai", "Halo", "Apa kabar?", "Ada orang di sana?", "Selamat pagi"],
         "responses": ["Halo!", "Hai!", "Salam!", "Bagaimana saya bisa membantu?"]},
        {"tag": "selamat_tinggal", "patterns": ["Selamat tinggal", "Sampai jumpa", "Dadah"],
         "responses": ["Selamat tinggal!", "Sampai jumpa!", "Semoga harimu menyenangkan!"]},
        {"tag": "terima_kasih", "patterns": ["Terima kasih", "Makasih", "Itu sangat membantu"],
         "responses": ["Sama-sama!", "Senang bisa membantu!", "Kapan saja!"]},
        {"tag": "dokter", "patterns": ["Saya butuh dokter", "Panggil dokter", "Saya butuh bantuan medis", "Saya butuh dokter umum"],
         "responses": ["Saya bisa membantu Anda menemukan dokter.", "Mari kita dapatkan bantuan medis untuk Anda.", "Silakan tunggu sementara saya menghubungkan Anda dengan dokter. Apakah Anda memerlukan spesialis tertentu?"]}
    ]
}

import json
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import random

# Simpan dataset ke file JSON
with open('intents_id.json', 'w') as outfile:
    json.dump(data, outfile)

# Memuat dataset
with open('intents_id.json') as file:
    intents = json.load(file)

# Memisahkan data menjadi input (patterns) dan output (tags)
patterns = []
tags = []
responses = {}

for intent in intents['intents']:
    for pattern in intent['patterns']:
        patterns.append(pattern)
        tags.append(intent['tag'])
    responses[intent['tag']] = intent['responses']

# Tokenize dan pad sequences
tokenizer = Tokenizer(num_words=2000)
tokenizer.fit_on_texts(patterns)
sequences = tokenizer.texts_to_sequences(patterns)
padded_sequences = pad_sequences(sequences, maxlen=20)

# Konversi tags menjadi one-hot encoding
tag_tokenizer = Tokenizer()
tag_tokenizer.fit_on_texts(tags)
tag_sequences = tag_tokenizer.texts_to_sequences(tags)
one_hot_tags = tf.keras.utils.to_categorical(tag_sequences)

# Membangun Model
model = Sequential()
model.add(LSTM(128, input_shape=(20, 1), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(64))
model.add(Dropout(0.2))
model.add(Dense(len(tag_tokenizer.word_index) + 1, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Melatih Model
padded_sequences = np.array(padded_sequences)
padded_sequences = np.reshape(padded_sequences, (padded_sequences.shape[0], padded_sequences.shape[1], 1))

history = model.fit(padded_sequences, np.array(one_hot_tags), epochs=200, batch_size=5, verbose=1)

# Fungsi untuk Menggunakan Model
def chat_response(text):
    sequence = tokenizer.texts_to_sequences([text])
    padded_sequence = pad_sequences(sequence, maxlen=20)
    padded_sequence = np.reshape(padded_sequence, (padded_sequence.shape[0], padded_sequence.shape[1], 1))
    predicted = model.predict(padded_sequence)
    tag = tag_tokenizer.index_word[np.argmax(predicted)]

    return random.choice(responses[tag])

# Contoh percakapan
print("User: Halo")
print(f"Sistem: {chat_response('Halo')}")

print("User: Saya butuh dokter")
print(f"Sistem: {chat_response('Saya butuh dokter')}")

print("User: Saya butuh dokter umum")
print(f"Sistem: {chat_response('Saya butuh dokter umum')}")

print("User: Terima kasih")
print(f"Sistem: {chat_response('Terima kasih')}")
```

```python
# Install library yang diperlukan
!pip install tensorflow pandas scikit-learn

# Import library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Dataset penyakit dan gejala
data = pd.DataFrame({
    'penyakit': [
        "Demam Berdarah", "Tifus", "Flu",
        "Diabetes", "Maag", "Anemia"
    ],
    'gejala': [
        "Demam tinggi", "Demam malam",
        "Batuk pilek", "Sering haus",
        "Sakit lambung", "Lemas dan pusing"
    ]
})

# Encoding penyakit
le = LabelEncoder()
data['penyakit_encoded'] = le.fit_transform(data['penyakit'])

# Tokenisasi teks
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(data['gejala'])
sequences = tokenizer.texts_to_sequences(data['gejala'])
padded_sequences = pad_sequences(sequences, maxlen=10, padding='post')  # Mengurangi panjang sequence untuk efisiensi

# Split data
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, data['penyakit_encoded'], test_size=0.2, random_state=42)

# Model deep learning
model = Sequential([
    Embedding(1000, 16, input_length=10),  # Sesuaikan panjang input sequence
    LSTM(32),  # Mengurangi jumlah LSTM units untuk model yang lebih ringan
    Dense(16, activation='relu'),
    Dense(len(le.classes_), activation='softmax')
])

# Menggunakan optimizer dengan learning rate yang lebih kecil
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training model dengan lebih banyak epoch
history = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))  # Meningkatkan jumlah epoch

# Evaluasi model
_, accuracy = model.evaluate(X_test, y_test)
print(f"Akurasi pada data validasi: {accuracy * 100:.2f}%")

# Fungsi chatbot
def chatbot_response(input_text):
    seq = tokenizer.texts_to_sequences([input_text])
    padded = pad_sequences(seq, maxlen=10, padding='post')
    prediction = model.predict(padded)
    predicted_class = np.argmax(prediction)
    confidence = prediction[0][predicted_class]
    response = le.inverse_transform([predicted_class])[0]
    return response, confidence

# Interaksi chatbot
while True:
    user_input = input("Anda: ")
    if user_input.lower() in ['exit', 'quit']:
        print("Chatbot: Terima kasih, semoga lekas sembuh!")
        break

    response, confidence = chatbot_response(user_input)
    print(f"Chatbot: Berdasarkan gejala yang Anda sebutkan, kemungkinan penyakit Anda adalah: {response} (keyakinan: {confidence * 100:.2f}%)")
```
